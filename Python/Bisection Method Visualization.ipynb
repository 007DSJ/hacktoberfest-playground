```python
import numpy as np
import matplotlib.pyplot as plt
#  +4x3 +3x1 -9x0
#  +7x7 -5x5 +3x3 +1x1 +1x0

expsm = input("Enter the expression : ")
terms = expsm.split()
terms 
tst = 2
sum ,a,b= 0,0,0
#for term in terms:
#    print(term)
coeff = list()
power = list()
sign = list()
for term in terms:
    coeff.append(term[1])
    power.append(term[3])
    if term[0] == '+':
        sign.append(1)
    else:
        sign.append(-1)
#for i in range(0,len(terms)):
#    sum+= int(sign[i]) * int(coeff[i]) * (int(tst) ** int(power[i]))
#print(sum)

#use recursive
#for root assumption using approximate values method

check = list()
for x in range(-200,200):
    sum = 0
    for i in range(0,len(terms)):
        sum += int(sign[i]) * int(coeff[i]) * (x ** int(power[i]))
    check.append(sum)

#for high precision appromixation
for c in range(0,len(check)):
    if check[c] < 0 and check[c+1] > 0:
        a = check[c]
        b = check[c+1]
rootlist = list()
iterations = input("Enter number of iterations : ")
for k in range(0,int(iterations)):
    x1 = (a+b)/2
    rootlist.append(x1)
    sum = 0
    for i in range(0,len(terms)):
        sum += int(sign[i]) * int(coeff[i]) * (x1 ** int(power[i]))
    if sum < 0:
        a = x1
    if sum > 0:
        b = x1
print(int(iterations) ,"th iteration = ",x1)
roots = np.array(rootlist)
plt.plot(roots, linestyle= 'dotted' ,c = 'r')
titles = "Approximate root value for "+expsm+" is "+str(x1)
plt.title(titles)
plt.xlabel('Iteration Number')
plt.ylabel('Assumed Root')
plt.show
```
